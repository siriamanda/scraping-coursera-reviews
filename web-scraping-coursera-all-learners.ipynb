```python
# Create data storage

pages = []
review = []
date = []
stars = []

# I am using a for loop to iterate through the pages and save the date and review information in the lists

url = 'https://www.coursera.org/learn/python/reviews?sort=recent'
for i in range(399):
    pg = url + '&page=' + str(i)
    r = requests.get(pg)
    soup = BeautifulSoup(r.text, parser = "html5lib")

    print("Status code: " + str(r.status_code))     # Ensure that we are able to download data from all pages

    # store response into "pages" list
    pages.append(r)

    # adding wait time:
    wait_time = randint(1,4)
    print("I will sleep for " + str(wait_time) + " second/s.")
    time.sleep(wait_time)
            
    for i in range(len(soup.select("div.reviewText > div"))):
        date.append(soup.select("div.dateOfReview.p-x-1s.m-b-0.text-secondary.font-xs")[i].get_text())
        review.append(soup.select("div.reviewText > div")[i].get_text())
        stars.append(soup.select("div._1qfi0x77 > div._1mzojlvw")[i].get_text())
```

```python
stars
```

```python
num_stars = []

for tag in stars:
        num_stars.append(tag.count('Filled Star'))
        
```

```python
num_stars
```

```python
# Create dataframe from lists

data_reviews_all = pd.DataFrame({"date": date,
                                 "review": review,
                                 "rating": num_stars})
```
